## 1.3 (3 points) Apply Principal Component Analysis (PCA) to the data of Xtrn_nm using sklearn.decomposition.PCA, and report the variances of projected data for the first five principal components in a table. Note that you should use Xtrn_nm instead of Xtrn.

```python
# Q1.3
def iaml01cw2_q1_3():
    pca_5 = PCA(n_components = 5)
    pc_5 = pca_5.fit_transform(Xtrn_nm)
    Xtrn_nm_cov = np.cov(pc_5.T)
    cov = []
    for i in range(len(Xtrn_nm_cov)):
        cov.append(Xtrn_nm_cov[i][i])
    return cov
#
iaml01cw2_q1_3()   # comment this out when you run the function
```

```python
# Q1.4
def iaml01cw2_q1_4():
    pca = PCA()
    pc = pca.fit_transform(Xtrn_nm)
    plt.plot(pca.explained_variance_ratio_.cumsum())
    plt.grid(True)
    plt.xlabel("K")
    plt.ylabel("Ratio")
    plt.title("Cumulative Explained Variance Ratio")
    plt.savefig('q1_4.png')
    plt.show()

iaml01cw2_q1_4()   # comment this out when you run the function
```

## 1.6 (5 points) Using Xtrn_nm, for each class and for each number of principal components K = 5,20,50,200, apply dimensionality reduction with PCA to the first sample in the class, reconstruct the sample from the dimensionality-reduced sample, and report the Root Mean Square Error (RMSE) between the original sample in Xtrn_nm and reconstructed one.
You should report this using a table, in which each row corresponds to class and each column corresponds to a value of K.

```python
# Q1.6
def iaml01cw2_q1_6():
    K = [5, 20, 50, 200]
    for i in K:
        print("K = ", i)
        pca = PCA(n_components=i)
        pc_k = pca.fit_transform(Xtrn_nm)

        pc_k_classes = split_classes(pc_k,Ytrn)
        pc_k_classes_split = i_class(pc_k_classes, Ytrn)

        for j in range(len(pc_k_classes_split)):
            first_sample = pc_k_classes_split[j].iloc[0,:-1]
            inverse_first_sample = pca.inverse_transform(first_sample)
            rmse = sqrt(mean_squared_error(classes_nm[j].iloc[0,:-1], inverse_first_sample))
            print("class {}: {:.3f}%".format(j, rmse*100))

iaml01cw2_q1_6()   # comment this out when you run the function
```

## 1.7 (4 points) Display the image for each of the reconstructed samples in a 10-by-4 grid, where each row corresponds to a class and each row column corresponds to a value of K = 5, 20, 50, 200.
Note that you should add Xmean to each reconstructed sample to display the corresponding image. Discuss your findings briefly.

```python
# Q1.7
def iaml01cw2_q1_7():
    K = [5, 20, 50, 200]
    flg, ax = plt.subplots(10,4, figsize=(20,40))
    for i in range(len(K)):
        print("K = ", K[i])
        pca = PCA(n_components=K[i])
        pc_k = pca.fit_transform(Xtrn_nm)

        pc_k = pca.inverse_transform(pc_k)
        Xtrn_k = pc_k + Xmean
        
        new_Xtrn_classes = split_classes(Xtrn_k, Ytrn)
        new_classes = i_class(new_Xtrn_classes, Ytrn)

        for j in range(len(new_classes)):
            ax[j,i].title.set_text("Label 0 Class {} with K {}".format(j,K[i]))
            ax[j,i].imshow(np.reshape(new_classes[j].iloc[0,:-1].values, (28,28)), 'gray_r')
    plt.savefig('q1_7.png')
#
iaml01cw2_q1_7()   # comment this out when you run the function
```
